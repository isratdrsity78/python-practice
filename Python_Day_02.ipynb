Function with **Parameter**

```python
def greet(name):
    print("Hello", name)

greet("Dristy")
greet("Israt")
```

**Function with Return**

```python
def add(a, b):
    return a + b

result = add(10, 20)

print(result)
```

**Even/Odd**

```python
def even_odd(n):
    if n % 2 == 0:
        return "Even"
    else:
        return "Odd"

print(even_odd(4))
```

**Student Result Checker**

```python
# Student Information
name = input("Enter Student Name: ")
roll = input("Enter Roll: ")

python_marks = int(input("Enter Python Marks: "))
math_marks = int(input("Enter Math Marks: "))
english_marks = int(input("Enter English Marks: "))

# Function
def calculate_total(py, math, eng):
    return py + math + eng

def calculate_average(total):
    return total / 3

def calculate_grade(avg):
    if avg >= 80:
        return "A+"
    elif avg >= 70:
        return "A"
    elif avg >= 60:
        return "B"
    elif avg >= 50:
        return "C"
    else:
        return "F"

# Function Call
total = calculate_total(python_marks, math_marks, english_marks)
average = calculate_average(total)
grade = calculate_grade(average)

# Output
print("\n------ Student Result ------")
print("Name :", name)
print("Roll :", roll)
print("Total :", total)
print("Average :", average)
print("Grade :", grade)
```
